# draw a plot for time and distance by using list comprehension 
* Distance=2*time


```python

import matplotlib.pyplot as q
x=[i for i in range(1,10)]
y=[i*2 for i in range(1,10)]
q.plot(x,y)
q.xlabel('x-time')
q.ylabel('y-distance')
q.title('time and distance')
q.show()
```

### Representing a sine wave
* A=sine(wt) ,A=Amplitude ,w=frequency ,t=time

```python
import numpy as np
import matplotlib.pyplot as pl
Time=np.arange(1,10,0.1)
Amplitude=np.sin(Time)
pl.plot(Time,Amplitude)
pl.title("sine wave A=sin(Wt)")
pl.xlabel('Time')
pl.ylabel('Amplitude')
pl.grid()
pl.show()
```
